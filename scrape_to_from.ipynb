```python
import pandas as pd
import datetime
import requests
from bs4 import BeautifulSoup
```

```python
def last20Years():
    # Create an empty list
    years_list = []

    # Get the current year
    current_year = datetime.datetime.now().year

    # Create a loop with starting value of (current year - 15) and end value of current year
    for i in range(current_year - 20, current_year + 1):
        # Append each year to the list
        years_list.append(i)

    # Return the list
    return years_list


# Call last20Years()
list_20years = last20Years()
```

```python
matching_str = (
    "/data-and-analysis/uk-aviation-market/airports/uk-airport-data/uk-airport"
)


def extract_links(url, string):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    links = []
    for link in soup.find_all("a"):
        if string in link.get("href"):
            links.append(link.get("href"))
    return links
```

```python
master_links = []
for year in list_20years:
    url_to_scrape = f"""https://www.caa.co.uk/data-and-analysis/uk-aviation-market/airports/uk-airport-data/uk-airport-data-{year}"""
    links = extract_links(url_to_scrape, matching_str)
    if len(links) > 1:
        master_links.append(links)
master_links
```

```python
def extract_links_text(url, string, file_to_check):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    links = []
    text = []
    for link in soup.find_all("a"):
        if string in link.get("href") and file_to_check in link.text:
            links.append(link.get("href"))
            text.append(link.text)
    return links, text
```

```python
def flatten_list(nested_list):
    return [item for sublist in nested_list for item in sublist]


master_links_flattened = flatten_list(master_links)
```

```python
def extract_links_for_specific_files(list_of_links, partital_file_name):
    csv_links = []
    csv_names = []

    for page in list_of_links:
        url = f"https://www.caa.co.uk/{page}"
        text_to_check_for = "/Download/"
        file_to_check_for = partital_file_name
        links, text = extract_links_text(
            url, text_to_check_for, file_to_check_for)
        csv_links.append(links)
        csv_names.append(text)
    return csv_links, csv_names


csv_links, csv_names = extract_links_for_specific_files(
    master_links_flattened, "Intl Air Pax Traffic Route Analysis (CSV"
)
```
